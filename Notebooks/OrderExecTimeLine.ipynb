# ITCH NASDAQ Order execution analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
from FSociety.Data.TimeLineSummary import timelines, ntimelines, stat, order_exec_analysis
```

```python
print(stat)
print(timelines)
print(ntimelines)
```

```python
def plot_dist(s, v):
    plt.figure(figsize=(20,10))
    plt.title(f'{stock} {s} {v}')
    for i in timelines[1:-1]:
        sns.distplot(statistics[i][s][v], label=ntimelines[timelines.index(i)], norm_hist=True, bins=50)
    plt.legend()
```

```python
stock= 'GOOGL'
year = '2017G'
day = 0
```

```python
statistics = order_exec_analysis(year, day, stock)
```

Distribucion entre la diferencia entre el precio de ejecucion y el siguiente precio para la ejecuciones de compras
